**In this tutorial, we demonstrate how to create a Northwest Atlantic Grid.**
<br>

Once the Gridtool Python library is installed, we are able to use the GridTool Application to generate custom grids, visualize them, and save them as netCDF files. In this tutorial, we cover the basics of the application and demonstrate the generation of a Northwest Atlantic grid. # Create the Grid

<div>
    <img src="../images/gridCenter.png" width="500" height="500">
</div>

<br>

**Let's create a grid located in the Northwest Atlantic and runs along the East Coast. Let's also give this grid a lambert conic conformal projection.**

1. Navigate to the Grid Tab. 
   - *Notice that a user is automatically placed in the **Center** sub-tab of the parent Grid Tab*
2. Specify a center longitude of **300 degrees**.
3. Navigate to the **Projection** sub-tab.
4. Set the **Projection** to **Lambert Conic Conformal**. 
   - *This is the projection of the grid, NOT the projection of the visualized grid in the plot window*
   - *If the grid projection and the plot (display) projection are not the same, the grid may appear distorted*
5. Specify a center longitude of **300 degrees** and a tilt of **-35 degrees**.
6. Navigate to the **Spacing** sub-tab.
7. Specify a **dx of 25** and a **dy of 35**
8. Select the **Make Grid** button.

# Plot the Grid

<div>
    <img src="../images/plotNWAtl.png" width="700" height="700">
</div>

<br>

**The plot projection only affects the projection of the image that visualizes the grid. When we made the grid above, a plot was automatically generated. However, the grid did not automatically fall within the default plot extent. First, we need to edit the plot extent.** 


1. Navigate to the Plot Tab. 
   - *Notice that a user is automatically placed in the **Projection** sub-tab of the parent Plot Tab*
2. Specify a plot projection of **Lambert Conformal Conic** (or feel free to keep this as Nearside Perspective).
   - *This ONLY changes the projection of the **visualized** grid, NOT the projection of the grid itself.*
3. Specify a center longitude of **300 degrees**.
4. Navigate to the Extent sub-tab.
5. Specify a **Longitude(x0)** of **-100 degrees**, a **Longitude(x1)** of **-35 degrees**, a **Latitude(y0)** of **15 degrees** and a **Latitude(y1)** of **65 degrees**.
6. Navigate to the Style sub-tab.
7. Specify the Title to your choosing.
8. Select the **Plot** button.

# Download the Grid
<div>
    <img src="../images/gridInfo.png" width="500" height="500">
</div>

<br>
<br>

**Once we are satisfied with the grid we've generated, we can proceed to download the Grid.**


1. Navigate to the **Grid Info** window Tab.
2. Double Check grid attributes (grid parameters, projection, extents, etc.). 
   - Notice that the grid cell area and grid cell angles have been automatically computed based on our specifications to the grid above.
3. Navigate to the **Local Files** window Tab.
4. Name the **Grid filename** as you wish.
5. Click the green **Download Grid** button.
   - The file will automatically be downloaded to the Downloads folder on your local machine.
